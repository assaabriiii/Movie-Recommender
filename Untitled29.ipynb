```python
import pandas as pd
```

```python
movies = pd.read_csv("/content/data/ml-latest-small/movies.csv")
ratings = pd.read_csv("/content/data/ml-latest-small/ratings.csv")
```

```python
ratings
```

```python
movies
```

```python
df = pd.merge(ratings, movies, on="movieId")
```

```python
df
```

```python
movie_matrix = df.pivot_table(index='userId', columns='title', values='rating')
```

```python
movie_matrix
```

```python
rating_count = df.groupby('title')['rating'].count()
rating_count
```

```python
popular_movies = rating_count[rating_count > 100].index
```

```python
movie_matrix = movie_matrix[popular_movies]
```

```python
movie_matrix
```

```python
my_profile_data = {
    'title': ['Matrix, The (1999)', 'Forrest Gump (1994)', 'Star Wars: Episode IV - A New Hope (1977)'],
    'rating': [5, 2, 5],
}
```

```python
my_ratings = pd.DataFrame(my_profile_data)
```

```python
my_ratings.to_csv('my_likes.csv', index=False)
```

```python
similar_movies = pd.DataFrame()
```

```python
for index, row in my_ratings.iterrows():
  my_movie = row['title']
  my_rating = row['rating']

  if my_movie in movie_matrix.columns:
    corr_scores = movie_matrix.corrwith(movie_matrix[my_movie])

    weighted_scores = corr_scores * (my_rating - 2.5)

    similar_movies = pd.concat([similar_movies, weighted_scores], axis=1)
```

```python
final_scores = similar_movies.sum(axis=1)
```

```python
final_scores = final_scores.sort_values(ascending=False)
```

```python
recommendations = final_scores.drop(my_ratings['title'], errors='ignore')
```
